#Data Analytics

```python
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['lama pengiriman'] = (df['Ship_Date'] - df['Order_Date']).dt.days
print(df[['Order_Date', 'Ship_Date', 'lama pengiriman']].head())
```

```python
conditions = [
    df['Ship_Mode'] == 'Standard Class',
    df['Ship_Mode'] == 'Second Class',
    df['Ship_Mode'] == 'Same Day',
    df['Ship_Mode'] == 'First Class'
]

values = [
    5 / df['lama pengiriman'],
    3 / df['lama pengiriman'],
    1 / df['lama pengiriman'],
    2 / df['lama pengiriman']
]

df['Shipping Performance'] = np.select(conditions, values, default=0)
df.loc[df['lama pengiriman'].isnull(), 'Shipping Performance'] = 1
df['Shipping Performance'] = df['Shipping Performance'] * 100
```

```python
lower_performance = df.loc[df['Shipping Performance'] < 100, ['Order_ID', 'Ship_Mode', 'Shipping Performance']]
print(lower_performance)
len(lower_performance)
(len(lower_performance)/len(df))*100
```

```python
lower_performance_count = len(lower_performance)
total_count = len(df)
lower_performance_percentage = (lower_performance_count / total_count) * 100
other_percentage = 100 - lower_performance_percentage

labels = ['Lower Performance', 'Others']
sizes = [lower_performance_percentage, other_percentage]
colors = ['red', 'blue']
explode = (0.1, 0)

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Lower Performance')
plt.show()
```
